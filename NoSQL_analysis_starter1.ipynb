- Search within 0.01 degree on either side of the latitude and longitude.
- Rating value must equal 5
- Sort by hygiene score

```python
# Define the degree search range
degree_search = 0.01

# Create the projection dictionary
projection = {'geocode.latitude':1,'geocode.longitude':1,'_id':0}

# Find documents within the specified projection
documents_cursor = establishments.find({},projection)

# Initialize a list to store documents with latitude_value and longitude_value
documents_with_coordinates = []

# Iterate through the cursor to find the first document within latitude and longitude
for document in documents_cursor:
    if 'geocode' in document and 'latitude' in document['geocode'] and 'longitute' in document['geocode']:
        documents_with_coordinates.append(document)

# Create empty list to store the result establishments
result_establishments_list = []

# Query to search within 0.01 degree on either side of the latitude and longitude       
for doc in documents_with_coordinates:
    latitude_value =doc['geocode']['latitude']
    longitude_value = doc['geocode']['longitude']

    query = {
         'geocode.latitude':{'$gte':latitude_value - degree_search, '$lte':latitude_value + degree_search},
         'geocode.longitude':{'$gte':latitude_value - degree_search, '$lte' : longitude_value + degree_search},
         'RatingValue':5    
    }    

# Sort the criteria by hygine score
sort = [('scores.Hygine',1)]

# Find establishment matching the query and sorted the hygine score
result_establishments = establishments.find(query).sort(sort)

# Append the results to the list
result_establishments_list.extend(list(result_establishments))
```

```python
# Convert result to Pandas DataFrame
establishments_df = pd.DataFrame(result_establishments_list)
establishments_df.head(5)
```
